```python
import numpy as np

from Pinsky_Rinzel import PinskyRinzel
```

```python
Voltages = np.arange(-85, 51, 15)
Ca_concentrations = np.arange(0.0, 2.1e-3, 0.5e-3)

for V in Voltages:
    for Ca in Ca_concentrations:
        PR = PinskyRinzel(exp_time=0.5, V=V, Ca=Ca)
        _ = PR.run(J=[0.0], times=[(0, 2)])
```
