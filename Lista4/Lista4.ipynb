c) Linearize o sistema de equações (1) e (2) para determinar a natureza do(s) ponto(s) de
equilı́brio. Para fazer isso, seja (v ∗ , w ∗ ) um ponto de equilı́brio do sistema. Considere
pequenas perturbações em torno deste ponto de equilı́brio: v = v ∗ + δv e w = w ∗ +
δw. Substitua estas expressões no sistema de equações e despreze os termos de ordem
quadrática ou superior em δv e δw. Você deve ficar com um sistema do tipo: <br>
que pode ser escrito na forma matricial, A matriz [ p r q s ] é chamada de matriz jacobiana do sistema no ponto de equilı́brio
(v ∗ , w ∗ ). É importante notar que cada ponto de equilı́brio tem uma matriz jacobiana
diferente. Quais os valores de p, q, r e s para cada um dos pontos de equilı́brio do
sistema?<br>
</div>

$ dv/dt = v - v³/3 - w + I $ <br>
$ dw/dt = 0.08(v + 0.7 - 0.8w) $ <br>
fazendo: <br>
$ v = v^{∗} + δv $, <br>
$ w = w^{∗} + δw $ and <br>
$ I = 0 $ <br>

$ => dv/dt = (v^{∗} + δv) - (v^{^{∗}} + δv)³/3 - (w^{∗} + δw) + 0 $ <br>
$ => dw/dt = 0.08((v^{∗} + δv) + 0.7 - 0.8(w^{∗} + δw)) $ <br>

$ => dv/dt = v^{∗} + δv - (v^{∗}³ + 3v^{∗}²δv + 3v^{∗}δv²+ δv³)/3 - w^{∗} + δw $ <br>
$ => dw/dt = 0.08(v^{∗} + δv + 0.7 - 0.8w^{∗} + 0.8δw) $ <br>

Calculando a Jacobiana: <br>
$ => J(δv, δw) =  \begin{bmatrix} dv_{δv}/dt & dv_{δw}/dt \\ dw_{δv}/dt & dw_{δw}/dt \end{bmatrix} = \begin{bmatrix} p & r \\ r & s \end{bmatrix} $ <br>

$ => dv_{δv}/dt = 1 - (3v^{∗} + 6v^{∗}δv + 3δv²)/3 = 1 - v^{∗} - 2v^{∗}δv - δv² $ <br>
$ => dv_{δw}/dt = -1 $ <br>
$ => dw_{δv}/dt = 0.08 $ <br>
$ => dw_{δw}/dt = 0.064 $ <br>

$ => J(δv, δw) =  \begin{bmatrix} 1 - v^{∗} - 2v^{∗}δv - δv² & -1 \\ 0.08 & 0.064 \end{bmatrix} $ <br>

usando: <br>
$ v^{∗} = -0.98 $ <br>

$ => dv_{δv}/dt =  1 - (-0.98) - 2 (-0.98) δv - δv²  = 1.98 + 1.96δv - δv²$ <br>
$ => J(δv, δw) =  \begin{bmatrix} 1.98 + 1.96δv - δv² & -1 \\ 0.08 & 0.064 \end{bmatrix} $ <br>

portanto: <br>

$ p = 1.98 + 1.96δv - δv² $ <br>
$ q = -1 $ <br>
$ r = 0.08 $ <br>
$ s = 0.064 $ <br>

```python
runner = FitzHughNagumo(exp_time=40)

voltages = [-0.80, -0.65, -0.64, 0]

volt_list = []

w_list = []

for v0 in voltages:
    V, n, i_inj_values = runner.run(J=[0], times=[(0, 40)], V0=v0, plot=False)
    volt_list.append(V)
    w_list.append(n)
```

```python
def plot_voltages(volt_list, label_list, exp_time):
    """Plot voltages for FitzHugh-Nagumo model.
    
    Parameters
    ----------
    volt_list : list
        List of voltages.
    label_list : list
        List of labels for voltages.
    exp_time : float
        Experiment time.
    """
    t = np.arange(0.0, exp_time, 1)

    plt.figure(figsize=(16, 10))
    plt.title("FitsHugh-Nagumo Neuron Voltages")
    color_map = ['r', 'g', 'b', 'c']

    for volt, color, label in zip(volt_list, color_map, label_list):
        plt.plot(t, volt, color, label=str(label))
    plt.legend()
```

(iii) Um gráfico do plano de fase w × v mostrando as nulclinas de v e w (em cores ou sı́mbolos diferentes) e as trajetórias correspondentes aos
quatro casos simulados (também em cores ou sı́mbolos diferentes).

```python
def plot_nullclines(volt_list, w_list, eq_points, label_list, exp_time):
    """Plot voltages for FitzHugh-Nagumo model.
    
    Parameters
    ----------
    volt_list : list
        List of voltages.
    label_list : list
        List of labels for voltages.
    exp_time : float
        Experiment time.
    """
    t = np.arange(0.0, exp_time, 1)

    plt.figure(figsize=(16, 10))
    plt.title("FitsHugh-Nagumo Neuron")
    color_map = ['r', 'g', 'b', 'c']
    nullcline_color_map = ['purple', 'orange', 'yellow', 'black']

    for volt, w, idx, color, null_color, label in zip(volt_list, w_list, eq_points, color_map, nullcline_color_map, label_list):
        plt.plot(t, volt, color, label=str(label))
        plt.plot(t, w, color)
        plt.plot(t[idx], volt[idx], "o", color=color)
        plt.plot(t[idx], w[idx], "o", color=color)
        plt.quiver(t[:-1], volt[:-1], volt[1:]-volt[:-1], w[1:]-w[:-1], scale_units='xy', angles='xy', scale=2, color=null_color, label=str(label))
        plt.quiver(t[:-1], w[:-1], volt[1:]-volt[:-1], w[1:]-w[:-1], scale_units='xy', angles='xy', scale=2, color=null_color)
    plt.legend()
```

```python
eq_points = []

for v, w in zip(volt_list, w_list):
    eq_points.append(find_equilibrium(v, w))

plot_nullclines(volt_list, w_list, eq_points, voltages, exp_time=40)
```

k) Construa a curva f-I do modelo de FitzHugh-Nagumo para I no intervalo 0 ≤ I ≤ 1. O
modelo é de tipo I ou tipo II? <br>

Podemos ver que o modelo aparenta ser do tipo II, pois com corrent baixas não há a presença de um trem de disparos.

</div>

```python
def get_number_of_action_potencial(V):
    """ Get number of action potentials.

    Parameters
    ----------
    V : array
        The Voltage vector.
    """
    V_threshold = V[V>40]
    asign = np.sign(V_threshold[:-1] - V_threshold[1:])
    signchange = ((np.roll(asign, 1) - asign) != 0).astype(int)
    return len(signchange[signchange == 1])/2
```

```python
current_inj = np.arange(0, 1, 0.2)
all_activation_potencials = []

for j in current_inj:
    V, w, i_inj_values = runner.run(J=[j], times=[(0, 40)])
    all_activation_potencials.append(get_number_of_action_potencial(V))
```

```python
plt.figure(figsize=(16, 10))
plt.plot(current_inj, np.array(all_activation_potencials))
plt.xlabel("Current (uA/cm^2)")
plt.ylabel("Number of action potentials")
plt.title("Number of action potentials vs current")
plt.show()
print(f"The max number of action potencials was {max(all_activation_potencials)} Hz")
print(f"The min number of action potencials was {min(all_activation_potencials)} Hz")
```
